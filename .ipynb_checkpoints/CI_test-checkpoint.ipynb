```stata
qui sum mpg, detail

scalar mean_mpg   = r(mean)
scalar se_mpg     = sqrt(r(Var)/r(N))
scalar t_cutoff   = invttail(r(N)-1, 0.025)
scalar upper_mean = mean_mpg + se_mpg * t_cutoff
scalar lower_mean = mean_mpg - se_mpg * t_cutoff

scalar list upper_mean lower_mean
```

```stata
qui sum mpg, detail

scalar var_mpg   = r(Var)
scalar chi_upper = invchi2tail(r(N)-1, 0.025)
scalar chi_lower = invchi2tail(r(N)-1, 0.975)
scalar upper_var = (r(N)-1) * var_mpg / chi_upper
scalar lower_var = (r(N)-1) * var_mpg / chi_lower

scalar list upper_var lower_var
```

```stata
qui sum foreign, detail

scalar mean_foreign = r(mean)
scalar se_foreign   = sqrt(r(Var)/r(N))
scalar t_cutoff     = invttail(r(N)-1, 0.025)
scalar upper_prop   = mean_foreign + se_foreign * t_cutoff
scalar lower_prop   = mean_foreign - se_foreign * t_cutoff

scalar list upper_prop lower_prop
```

```stata
matrix result_mean = e(b)
scalar mean_corr   = result_mean[1,1]
matrix result_se   = e(se)
scalar se_corr     = result_se[1,1]

scalar hyp_value = -0.7
scalar bs_test   = (mean_corr - hyp_value) / se_corr
scalar p_two     = 2 * normal(-1 * abs(bs_test))

display
display "Table: two-tail bootstrap test result"
display "---------------------------------------"
display "the bs test statistic is " bs_test
display "the two-tail p-vaplue is " p_two
display "---------------------------------------"

if p_two < 0.05{
    display "H0 is rejected since p value is less than sig level."
}
else if p_two >= 0.05{
    display "H0 is not rejected since p value is larger than sig level."
}
```

```stata
matrix result_mean = e(b)
scalar mean_corr   = result_mean[1,1]
matrix result_se   = e(se)
scalar se_corr     = result_se[1,1]

scalar hyp_value = -0.7
scalar bs_test   = (mean_corr - hyp_value) / se_corr
scalar p_left    = normal(bs_test)

display
display "Table: left-tail bootstrap test result"
display "---------------------------------------"
display "the bs test statistic  is " bs_test
display "the left-tail p-vaplue is " p_left
display "---------------------------------------"

if p_left < 0.05{
    display "H0 is rejected since p value is less than sig level."
}
else if p_left >= 0.05{
    display "H0 is not rejected since p value is larger than sig level."
}
```

```stata
matrix result_mean = e(b)
scalar mean_corr   = result_mean[1,1]
matrix result_se   = e(se)
scalar se_corr     = result_se[1,1]

scalar hyp_value = -0.7
scalar bs_test   = (mean_corr - hyp_value) / se_corr
scalar p_right   = 1 - normal(bs_test)

display
display "Table: left-tail bootstrap test result"
display "---------------------------------------"
display "the bs test statistic   is " bs_test
display "the right-tail p-vaplue is " p_right
display "---------------------------------------"

if p_right < 0.05{
    display "H0 is rejected since p value is less than sig level."
}
else if p_right >= 0.05{
    display "H0 is not rejected since p value is larger than sig level."
}
```
