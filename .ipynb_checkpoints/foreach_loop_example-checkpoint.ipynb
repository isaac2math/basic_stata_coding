```stata
gen lprice = log(price)

twoway hist lprice || kdensity lprice, title("histogram and density plot for log(price)")
```

```stata
qui twoway hist price || kdensity price, name(g1, replace) nodraw title("histogram and density for price")
qui twoway hist lprice || kdensity lprice, name(g2, replace) nodraw title("histogram and density for log(price)")

graph combine g1 g2, col(1) name(combined, replace)
```

```stata
tabstat price lprice, stat(skew kurt)
```

```stata
gen lmpg = log(mpg)

qui twoway hist mpg || kdensity mpg, name(g3, replace) nodraw title("histogram and density for mpg")
qui twoway hist lmpg || kdensity lmpg, name(g4, replace) nodraw title("histogram and density for log(mpg)")

graph combine g3 g4, col(1) name(combined, replace)
```

```stata
tabstat mpg lmpg, stat(skew kurt)
```

```stata
gen lmpg2 = lmpg^2

reg lprice lmpg lmpg2

qui estat ic

matrix M_IC = r(S)
scalar aic  = M_IC[1, 5]
scalar bic  = M_IC[1, 6]

display "the aic of this model is " aic
display "the bic of this model is " bic
```

```stata
gen lmpg3 = lmpg^3

reg lprice lmpg lmpg2 lmpg3

qui estat ic

matrix M_IC = r(S)
scalar aic  = M_IC[1, 5]
scalar bic  = M_IC[1, 6]

display "the aic of this model is " aic
display "the bic of this model is " bic
```

```stata
gen lmpg4 = lmpg^4

reg lprice lmpg lmpg2 lmpg3 lmpg4

qui estat ic

matrix M_IC = r(S)
scalar aic  = M_IC[1, 5]
scalar bic  = M_IC[1, 6]

display "the aic of this model is " aic
display "the bic of this model is " bic
```

```stata
reg lprice lmpg lmpg2

predict yhat, xb
```

```stata
twoway scatter lprice lmpg, mcolor(green) || scatter yhat lmpg, mcolor(red)
```

```stata
sysuse auto, clear

foreach i of numlist 1/4 {
    
    if `i' == 1{
        
        gen lprice = log(price)
        gen lmpg   = log(mpg)
        
    }
    
    if `i' != 1{
        
        gen lmpg`i' = log(mpg)^`i'
        
    }
    
    qui reg lprice lmpg*
    estimate store M_`i'
    
    predict yhat_`i', xb
    
    qui estat ic
    matrix M_IC_`i' = r(S)
    scalar aic_`i'  = M_IC_`i'[1, 5]
    scalar bic_`i'  = M_IC_`i'[1, 6]
    
  } 

estout M_*, cells(b(star fmt(3)) se(par fmt(3))) stats(r2 aic bic df_r, labels(R^2 AIC BIC df)) legend title("Table 1: polynomial regression comparison table")

twoway (scatter lprice lmpg, mcolor(green)) (scatter yhat_2 lmpg, mcolor(red)), title("Figure 1 : 2nd order predicted and actual value of Y")
```

```stata
twoway (scatter yhat_1 lmpg, mcolor(red)) (scatter yhat_2 lmpg, mcolor(blue)) (scatter yhat_3 lmpg, mcolor(black)) (scatter yhat_4 lmpg, mcolor(green)), legend(order(1 "1st order" 2 "2nd order" 3 "3rd order" 4 "4th order")) title("prediceted polynomial regressions with different orders")
```
